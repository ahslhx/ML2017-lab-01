```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split

%matplotlib inline

LEARNING_RATE = 0.1
NUM_ITERATIONS = 500
PATH = "./data/australian_scale.txt"

#load data
def get_data(path):
    data = load_svmlight_file(path)
    return data[0],  data[1]

#compute hinge loss
def compute_hinge_loss(X,y,W):
    n = X.shape[0]
    zero = np.zeros((n,1))
    pred = np.dot(X,W)
    p = X.dot(W)
    hinge_loss = np.max([zero,1 - y*pred],axis=0)
    return hinge_loss

#compute loss
def compute_loss(X,y,W,C,lambdaa):
    n = X.shape[0]
    hinge_loss = compute_hinge_loss(X,y,W)
    loss = np.sum(hinge_loss)*C/n +  np.dot(W,W.transpose())/2*lambdaa
    return loss[0][0]

#cmpute gradient
def compute_gradient(X,y,W,C,lambdaa):
    hinge_loss = compute_hinge_loss(X,y,W)
    indicator = np.zeros((X.shape[0],1))
    indicator[np.nonzero(hinge_loss)] = 1
    gradient = -np.dot(X.transpose(),y*indicator).sum(axis=1).reshape((X.shape[1], 1))*C/X.shape[0] + W*lambdaa
    return gradient

#compute accuracy
def compute_accuracy(X,y,W,threshold):
    n = X.shape[0]
    pred = np.dot(X,W)
    pred[pred <= threshold] = -1
    pred[pred > threshold] = 1
    count = pred*y
    count[count < 0] = 0
    accuracy = 1.0/n*np.count_nonzero(count)
    return accuracy

def test():
    path = PATH
    X,y = get_data(path)
    X0 = np.ones((len(y),1))
    X = X.toarray()
    X = np.column_stack((X0,X))
    y = y.reshape((len(y), 1))
    #print(X.shape)

    #divide dataset
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.3)
    num_features = X_train.shape[1]
    num_train = X_train.shape[0]
    num_validation = X_validation.shape[0]
    #initial parameters
    learning_rate = LEARNING_RATE
    num_iterations = NUM_ITERATIONS
    lambdaa = 0.1
    threshold = 0.5
    C = 1.0
    #initial W in three ways
    initial_W_zero = np.zeros((num_features,1))
    initial_W_random = np.random.random((num_features,1))
    initial_W_normal = np.random.normal(size=(num_features,1))
    
    lr_arr = np.array([0.01,0.05,0.1,0.2,0.3])
    color = np.array(['r','g','b','y','c'])
    lamda_arr = np.array([0.01,0.05,0.1,0.25,0.5])
    plt.figure(1)
    for j in np.arange(lr_arr.size):
        loss_train = np.zeros((num_iterations,1))
        W = initial_W_normal
        for i in range(0,num_iterations):
            loss_train[i] = compute_loss(X_train,y_train,W,C,lambdaa)
            G = compute_gradient(X_train,y_train,W,C,lambdaa)
            W = W - lr_arr[j]*G
        label = 'lr = ' + str(lr_arr[j])
        plt.plot(loss_train,c=color[j],label=label)
    plt.legend(loc='best')
    plt.xlabel("Iteration")
    plt.ylabel("Loss")
    plt.title("Linear Classification")
    
    plt.figure(2)
    for j in np.arange(lamda_arr.size):
        accuracy_train = np.zeros((num_iterations,1))
        W = initial_W_normal
        for i in range(0,num_iterations):
            accuracy_train[i] = compute_accuracy(X_train,y_train,W,threshold)
            G = compute_gradient(X_train,y_train,W,C,lamda_arr[j])
            W = W - learning_rate*G
        label = 'lambda = ' + str(lamda_arr[j])
        plt.plot(accuracy_train,c=color[j],label=label)
    plt.legend(loc='best')
    plt.xlabel("Iteration")
    plt.ylabel("Accuracy")
    plt.title("Linear Classification")
    
    threshold_arr = np.array([0.3,0.4,0.5,0.6,0.7])
    plt.figure(3)
    for j in np.arange(threshold_arr.size):
        accuracy_train = np.zeros((num_iterations,1))
        W = initial_W_normal
        for i in range(0,num_iterations):
            accuracy_train[i] = compute_accuracy(X_train,y_train,W,threshold_arr[j])
            G = compute_gradient(X_train,y_train,W,C,lambdaa)
            W = W - learning_rate*G
        label = 'threshold = ' + str(threshold_arr[j])
        plt.plot(accuracy_train,c=color[j],label=label)
    plt.legend(loc='best')
    plt.xlabel("Iteration")
    plt.ylabel("Accuracy")
    plt.title("Linear Classification")
    plt.show()
    
def main():
    path = PATH
    X,y = get_data(path)
    X0 = np.ones((len(y),1))
    X = X.toarray()
    X = np.column_stack((X0,X))
    y = y.reshape((len(y), 1))
    #print(X.shape)

    #divide dataset
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.3)
    num_features = X_train.shape[1]
    num_train = X_train.shape[0]
    num_validation = X_validation.shape[0]
    #initial W in three ways
    initial_W_zero = np.zeros((num_features,1))
    initial_W_random = np.random.random((num_features,1))
    initial_W_normal = np.random.normal(size=(num_features,1))
    
   #initial parameters
    learning_rate = LEARNING_RATE
    num_iterations = NUM_ITERATIONS
    W = initial_W_normal
    lambdaa = 0.1
    threshold = 0.5
    C = 1.0
    
    loss_train = np.zeros((num_iterations,1))
    loss_validation = np.zeros((num_iterations,1))
    accuracy_train = np.zeros((num_iterations,1))
    accuracy_validation = np.zeros((num_iterations,1))
    
    #perform gradient descent to learn W and update W  
    for i in range(0,num_iterations):
        loss_train[i] = compute_loss(X_train,y_train,W,C,lambdaa)
        loss_validation[i] = compute_loss(X_validation,y_validation,W,C,lambdaa)
        accuracy_train[i] = compute_accuracy(X_train,y_train,W,threshold)
        accuracy_validation[i] = compute_accuracy(X_validation,y_validation,W,threshold)
        #compute G
        G = compute_gradient(X_train,y_train,W,C,lambdaa)
        #update W
        W = W - learning_rate*G

        
    # plot
    fig = plt.figure(1)
    ax1 = fig.add_subplot(111)
    ax1.plot(loss_train,label="Training Loss")
    ax1.plot(loss_validation,label="Validation Loss")
    ax1.set_ylabel("Loss")
    ax1.set_xlabel("Iteration")
    ax1.legend(loc=4)
    ax2 = ax1.twinx() 
    ax2.plot(accuracy_train,label="Training Accuracy",color = 'r')
    ax2.plot(accuracy_validation,label="Validation Accuracy",color = 'b')
    ax2.legend(loc=1)
    ax2.set_ylabel("Accuracy")
    ax2.set_title("Linear Classification")
    plt.show()
    
if __name__ == "__main__":
    main()
```

```python
test()
```
